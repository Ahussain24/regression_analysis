```python
import numpy as np
import  matplotlib.pyplot as plt 
```

```python
plt.style.use('seaborn')
```

```python
np.random.seed(42)
```

```python
#generate dataset 

sample_size = 1000
m_true = 2.5
b_true = 1

X = np.random.uniform(low = 0,high = 21 ,size = sample_size)
noise = np.random.normal (loc = 0, scale = 5, size = sample_size)
y = m_true * X + b_true + noise 
```

```python
plt.scatter(X,y, color='red', label="regresion data")
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Data for linear regression')
plt.show()
```

```python
#use gradient descent to fit a line through the data
```

```python
W = 0
b = 0 
```

```python

y_pred = np.dot(X,W) + b
```

```python
plt.scatter(X,y, color='red', label="regresion data")
plt.plot(X,y_pred, color='blue', label='random_fit line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Data for linear regression')
plt.show()
```

```python
def mse(y,yhat) : 
    
    return np.mean(( yhat - y )**2) 
```

```python
def initialize_parameters() :  
    
    W = 0
    b = 0 
    
    return W, b
```

```python
def gradientDescent(X,y, lr = 0.01, max_epoch = 1000):


    
    costs = []
    
    W, b = initialize_parameters()
    
    for epoch in range(1, max_epoch+1) : 

        y_pred = X*W + b

        error = y_pred - y 

        loss = mse(y,y_pred)

        costs.append(loss)

        if not epoch % 10 :
            print(f"Epoch {epoch}  loss {loss} lr {lr}")

        dw = (1/len(y)) * np.dot(  X , error )
        db = (1/len(y)) *  np.sum( error) 

        W = W - lr * dw 
        b = b - lr * db 
    
    return costs, W,b 
```

```python
costs,W,b  = gradientDescent(X, y)
```

```python
plt.plot(list(range(1,len(costs)+1)), costs, color='red', label = 'loss wrt epoch') 
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend() 
plt.show()
```

```python
result = np.dot(X,W) + b
```

```python
plt.scatter(X,y, color='red', label="regresion data")
plt.plot(X,result, color='blue', label='random_fit line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Data for linear regression')
plt.show()
```

### RIDGE REGRESSION

```python
def ridgeGradientDescent(X,y, lambd, learning_rate = 0.01, max_epoch = 1000) : 
    
    W, b = initialize_parameters() 
    costs = [] 
    
    for epoch in range(1, max_epoch) : 
        
        pred = np.dot(X,W) + b 
        
        error = pred - y 
        
        loss = mse(y,pred) 
        
        costs.append(loss) 
        
        if not epoch % 10 : 
            print(f"Epoch {epoch} Loss {loss} ")
        
        dw = (1/len(y)) * np.dot(X, error) + (2 * lambd * (W**2) ) 
        db = (1/len(y)) * np.sum(error)
        
        W -= learning_rate * dw 
        b -= learning_rate * db
        
    return costs, W,b
    
```

```python
costs_ridge, W_ridge, b_ridge = ridgeGradientDescent(X,y,lambd=0.1)
```

```python
print(W_ridge, b_ridge)
print(output.coef_, output.intercept_)
print(W,b)
```

```python
plt.scatter(X,y, color='red', label="regresion data")
plt.plot(X,np.dot(X,W_ridge)+b_ridge, color='blue', label='random_fit line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Data for linear regression')
plt.show()
```

### LASSO REGRESSION 

```python
def lassoGradientDescent(X,y, lambd, learning_rate = 0.01, max_epoch = 1000) : 
    
    W, b = initialize_parameters() 
    costs = [] 
    
    for epoch in range(1, max_epoch) : 
        
        pred = np.dot(X,W) + b 
        
        error = pred - y 
        
        loss = mse(y,pred) 
        
        costs.append(loss) 
        
        if not epoch % 10 : 
            print(f"Epoch {epoch} Loss {loss} ")
        
        dw = (1/len(y)) * np.dot(X, error) + (lambd * np.abs(W) ) 
        db = (1/len(y)) * np.sum(error)
        
        W -= learning_rate * dw 
        b -= learning_rate * db
        
    return costs, W,b
```

```python
costs_lasso, W_lasso, b_lasso = lassoGradientDescent(X,y,lambd=1)
```
